# Question 5.) What are iterators in Python and how do they differ from iterables?

Answer 5.)

Iterators in Python are objects that allow you to traverse through all the elements of a collection (like a list or a tuple) one at a time. They maintain a state to keep track of where they are during iteration.


Iterables, on the other hand, are objects that can return an iterator. These include collections like lists, tuples, and strings. You can use the iter() function to get an iterator from an iterable.

Differences:

# Definition:

Iterable: Any object that can return an iterator.

Iterator: An object with a state that can iterate over elements.

# Implementation:

Iterable: Implements the __iter__() method.

Iterator: Implements both the __iter__() and __next__() methods.

# Usage:

Iterable: Can be looped over directly using a for loop.

Iterator: Needs to be explicitly fetched using next().

# State:

Iterable: Does not maintain state.

Iterator: Maintains state during iteration.

```python
# Iterable
my_list = [1, 2, 3]
for item in my_list:
    print(item)  # Outputs: 1, 2, 3

# Iterator
my_iterator = iter(my_list)
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
```

# Question 6.) Explain the concept of generators in Python and how they are defined?

Answer 6.) Generators are a special type of iterator in Python, used to iterate over a sequence of values without creating the entire sequence in memory at once. This makes them more memory-efficient for large datasets.

Concept: A generator function is defined using the def keyword, but instead of returning a single value, it yields a series of values using the yield keyword. Each call to the generator function resumes where it left off and continues until the sequence is exhausted.


In this example, countdown is a generator that yields a sequence of numbers from n down to 1. Each time yield n is executed, the function's state is preserved, allowing it to resume from the same point on the next call.

When you iterate over the generator using a for loop, it keeps yielding the next value in the sequence until there are no more values left. Generators are particularly useful for handling large data streams or infinite sequences, as they compute values on-the-fly.

```python
# Defining a Generator:

def countdown(n):
    while n > 0:
        yield n
        n -= 1


#Using a Generator:

for number in countdown(5):
    print(number)

# Output:
# 5
# 4
# 3
# 2
```

# Question.7) What are the advantages of using generators over regular functions?

Answer 7.) Generators can be real game-changers for your code. Hereâ€™s why:


Memory Efficiency: Generators produce items one at a time and only when asked, which means they don't load the entire sequence into memory. Ideal for handling large datasets or streams.


Lazy Evaluation: Values are computed on-the-fly as you iterate, so they are only generated when needed. This can improve performance by avoiding the overhead of generating all items up front.


Simpler Code: Generators allow you to write iterators in a clear, concise manner using the yield statement, reducing the complexity compared to maintaining state manually.


Infinite Sequences: Generators can represent infinite sequences, like a stream of Fibonacci numbers, without running out of memory.

```python
# example 1

def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
for _ in range(10):
    print(next(fib))  # Outputs the first 10 Fibonacci numbers
```

# Question 9.) Explain the purpose and usage of the `map()` function in Python.

Answer 9.) The map() function in Python applies a given function to all the items in an iterable (like a list or tuple) and returns a map object (an iterator) with the results. It's a convenient way to transform items in an iterable without using a for loop.

Purpose
The main purpose of map() is to apply a specific function to each item of an iterable, which can simplify and make your code more readable. It's especially useful for performing operations on large datasets or when you need to apply a common operation to all items in a collection.

```python
map(function, iterable)

# Example Let's say you want to square each number in a list:

# Regular function
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]

# example 2
# You can also use a lambda function to make it even more concise:


numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]

```

**Question 1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list?**



Hereâ€™s a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list:



```python
def sum_of_even_numbers(numbers):
    # Initialize a variable to store the sum of even numbers
    total = 0

    # Iterate through the list of numbers
    for num in numbers:
        # Check if the number is even
        if num % 2 == 0:
            total += num  # Add the even number to the total

    return total  # Return the sum of even numbers
```

**Example Usage:**



```python
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)  # Output: 12
```

**Explanation**

The function sum_of_even_numbers initializes a variable total to store the sum of even numbers.
It then loops through the list of numbers, checking if each number is even (i.e., divisible by 2).
If the number is even, it adds that number to total.
Finally, the function returns the sum of all even numbers.

**Question 4.Write a Python function that checks if a given number is prime or not from 1 to 200?**



 Python function that checks if a given number is prime or not within the range of 1 to 200

```python
def is_prime(n):
    # Return False for numbers less than 2 as they are not prime
    if n < 2:
        return False

    # Check if the number has any divisors other than 1 and itself
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False  # If a divisor is found, the number is not prime

    return True  # If no divisors are found, the number is prime
```

**Example Usage:**


```python
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
```

**Question 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?**



```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms  # Total number of terms to generate
        self.current_term = 0       # Current term index
        self.a, self.b = 0, 1       # First two numbers in the Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.num_terms:
            raise StopIteration  # Stop iteration when the limit is reached

        # Return the next Fibonacci number
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            # Calculate the next Fibonacci number
            next_fib = self.a + self.b
            self.a, self.b = self.b, next_fib
            self.current_term += 1
            return next_fib
```

**Example Usage:**


```python
# Create an instance of FibonacciIterator with 10 terms
fib = FibonacciIterator(10)

# Iterate through the Fibonacci sequence
for number in fib:
    print(number)
```

**Explanation**
The FibonacciIterator class takes num_terms as an argument, which specifies how many terms of the Fibonacci sequence to generate.
The __iter__() method allows the object to be used as an iterator.
The __next__() method generates the next Fibonacci number in the sequence. It raises a StopIteration exception when the specified number of terms is reached.
The first two Fibonacci numbers (0 and 1) are handled separately, and subsequent numbers are generated by summing the two preceding values (self.a + self.b)

**Question. 6 Write a generator function in Python that yields the powers of 2 up to a given exponent?**

```python
def powers_of_two(max_exponent):
    # Iterate from 0 to max_exponent
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent  # Yield 2 raised to the power of exponent
```

Example Usage:


```python
# Create a generator for powers of 2 up to exponent 5
for power in powers_of_two(5):
    print(power)
```

Hereâ€™s a Python generator function that reads a file line by line and yields each line as a string:



```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Yield each line without leading/trailing whitespace
```

Example Usage:

The result is converted back to a list and stored in fahrenheit_temperatures.

```python
# Define a function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

# Given string
input_string = "Hello, World!"

# Use filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(filtered_string)
```
