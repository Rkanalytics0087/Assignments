# Question 4.
Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
n_samples = 100
categories = ['A', 'B', 'C']
data = {
    'Category': np.random.choice(categories, size=n_samples),
    'Value': np.random.randint(1, 100, size=n_samples)
}
df = pd.DataFrame(data)

# Visualize distribution
plt.figure(figsize=(8, 6))
for category in categories:
    subset = df[df['Category'] == category]
    plt.hist(subset['Value'], alpha=0.7, label=category)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Values across Categories')
plt.legend()
plt.show()
```

# Question 5.
Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset usina a heatmap?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data with correlated features
np.random.seed(42)
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = 2 * x1 + np.random.randn(n_samples) * 0.5  # x2 is correlated with x1
x3 = -x1 + np.random.randn(n_samples) * 0.3 # x3 is negatively correlated with x1
x4 = np.random.rand(n_samples) # x4 is independent of the other features

data = {'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4}
df = pd.DataFrame(data)


# Visualize the correlation matrix using a heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

#**Question 3**
Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]?

Import necessary libraries.
Create a bar chart using Bokeh.


```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.models import ColumnDataSource

# Output the plot directly in a Jupyter Notebook
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Creating a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Setting up the figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           toolbar_location=None, tools="")

# Adding bar glyphs to the plot
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="orange")

# Customizing the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Display the plot
show(p, notebook_handle=True)
```

**Explanation:**
x_range=fruits sets the x-axis to the categories defined in fruits.
vbar() is used to draw vertical bars, with x specifying the categories and top indicating the height of each bar.
The ColumnDataSource makes it easy to manage the data within Bokeh

#**Question 4**.

Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np. random. randn (1000)
hist, edges = np.histogram(data_hist, bins=30)?

Import the necessary libraries.
Create the histogram plot with Bokeh

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook

# Generating the data
data_hist = np.random.randn(1000)

# Creating the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Bokeh plot setup
output_notebook()  # This will output the plot directly in a Jupyter Notebook

p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency')

# Adding the quad glyphs to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white")

# Show the plot
show(p, notebook_handle=True)
```

**Explanation:**
np.random.randn(1000) generates 1000 random data points from a normal distribution.
np.histogram(data_hist, bins=30) creates the histogram data.
The quad glyph in Bokeh is used to draw the histogram bars

#**Question 5**.

Create a Bokeh heatmap using the provided dataset.
data_heatmap = np. random. rand (10, 10)
x = np. linspace(0, 1, 10)
y = np. linspace(0, 1, 10)
xx, yy = np .meshgrid (x, y)?

Import necessary libraries.
Create a heatmap plot using the image glyph in Bokeh

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook

# Output the plot directly in a Jupyter Notebook
output_notebook()

# Generating the dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Setting up the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None, tools="")

# Adding the heatmap image
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette="Viridis256")

# Customizing the plot
p.grid.visible = False

# Display the plot
show(p, notebook_handle=True)
```

**Explanation**:
np.random.rand(10, 10) generates a 10x10 matrix of random values between 0 and 1.
The image() function in Bokeh is used to render the 2D data array as a heatmap.
The palette parameter defines the color scheme (e.g., Viridis256)
